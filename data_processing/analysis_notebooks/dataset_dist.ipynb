```python
# ==========================================
# UNSW-NB15 Original Dataset Distribution
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# 1. Load Datasets
# ------------------------------
train_df = pd.read_csv("UNSW_NB15_training-set.csv")
test_df = pd.read_csv("UNSW_NB15_testing-set.csv")

print("Training Shape:", train_df.shape)
print("Testing Shape:", test_df.shape)


# ------------------------------
# 2. Binary Distribution (0=Normal, 1=Attack)
# ------------------------------
print("\n========== BINARY DISTRIBUTION (TRAIN) ==========")
binary_train = train_df["label"].value_counts()
binary_train_percent = train_df["label"].value_counts(normalize=True) * 100

binary_summary_train = pd.DataFrame({
    "Count": binary_train,
    "Percentage": binary_train_percent.round(2)
})
binary_summary_train.index = binary_summary_train.index.map({0: "Normal", 1: "Attack"})
print(binary_summary_train)


print("\n========== BINARY DISTRIBUTION (TEST) ==========")
binary_test = test_df["label"].value_counts()
binary_test_percent = test_df["label"].value_counts(normalize=True) * 100

binary_summary_test = pd.DataFrame({
    "Count": binary_test,
    "Percentage": binary_test_percent.round(2)
})
binary_summary_test.index = binary_summary_test.index.map({0: "Normal", 1: "Attack"})
print(binary_summary_test)


# ------------------------------
# 3. Original 9-Class Distribution (attack_cat)
# ------------------------------
print("\n========== MULTI-CLASS DISTRIBUTION (TRAIN) ==========")
multi_train = train_df["attack_cat"].value_counts()
multi_train_percent = train_df["attack_cat"].value_counts(normalize=True) * 100

multi_summary_train = pd.DataFrame({
    "Count": multi_train,
    "Percentage": multi_train_percent.round(2)
})
print(multi_summary_train)


print("\n========== MULTI-CLASS DISTRIBUTION (TEST) ==========")
multi_test = test_df["attack_cat"].value_counts()
multi_test_percent = test_df["attack_cat"].value_counts(normalize=True) * 100

multi_summary_test = pd.DataFrame({
    "Count": multi_test,
    "Percentage": multi_test_percent.round(2)
})
print(multi_summary_test)


# ------------------------------
# 4. Combined Distribution (Train + Test)
# ------------------------------
full_df = pd.concat([train_df, test_df])

print("\n========== FULL DATASET DISTRIBUTION ==========")
full_multi = full_df["attack_cat"].value_counts()
full_multi_percent = full_df["attack_cat"].value_counts(normalize=True) * 100

full_summary = pd.DataFrame({
    "Count": full_multi,
    "Percentage": full_multi_percent.round(2)
})
print(full_summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# load one of your CSVs – change path if needed
df = pd.read_csv("UNSW_NB15_grouped_training_class.csv")

# assume the column holding the class labels is named 'label' or similar
label_col = "label"           # adjust to your actual column name
counts = df[label_col].value_counts()

# plot
fig, ax = plt.subplots(figsize=(6,6))
counts.plot.pie(
    autopct="%1.1f%%",
    startangle=90,
    textprops={"fontsize": 12},
    labels=[f"{lbl} ({cnt})" for lbl, cnt in counts.items()],
    ax=ax
)
ax.set_ylabel("") 
ax.set_title("Class distribution")
plt.tight_layout()
plt.show()
```
